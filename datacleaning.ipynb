## Data Cleaning

```python
marketing_data['Age'] = 2021-marketing_data.Year_Birth
marketing_data.head(2)
```

```python
# marketing_data['Marital_Status'].groupby(marketing_data['Marital_Status']).count()/marketing_data,shape[0]
```

```python
filter1 = marketing_data.loc[marketing_data['Marital_Status'].isin(['Divorced','Married','Single','Together','Widow']),:]
filter1.shape
```

```python
filter2 = filter1.loc[marketing_data['Age']<= 100, :]
filter2.shape
```

```python
filter3 = filter2.loc[filter2['Income'] != 666666,:]
filter3.shape
```
